```python
import folium
from folium.plugins import HeatMap
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read(dataset):
    """ imports the dataset into pandas
    """
    df = pd.read_csv(dataset)
    return df
```

```python
def generateBaseMap(default_location=[40.693942, -73.985880]):
    """ takes a location and generates a base map
    """
    base_map = folium.Map(location=default_location)
    return base_map
```

```python
def heatmap(df):
    """ creates a heatmap of the data
    """
    base_map = generateBaseMap()
    HeatMap(data=df[['latitude', 'longitude', 'price']].groupby( ['latitude' , 'longitude']).mean().reset_index().values.tolist(), radius=8,max_zoom=13).add_to(base_map)
    base_map.save('index.html')
    return base_map
```

```python
df = read('listings.csv')
```

```python
heat_map = heatmap(df)
heat_map
```
